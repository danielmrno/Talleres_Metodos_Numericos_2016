# Ejercicio 5.1

```python
import math
from tabulate import tabulate 
import matplotlib.pyplot as plt
%matplotlib inline

xReal=-1.405
a, b, c, xl, xu, xr, x1, x2=-0.5, 2.5, 4.5, 5.0, 10.0, 0.0, 0.0, 0.0
contador, x=1, 0
tabla, aux, funcion, raices, aprox=[], [],[], [],[]

def cuadratica():
    x1=(-b+math.sqrt(math.pow(b,2)-4*a*c))/(2*a)
    x2=(-b-math.sqrt(math.pow(b,2)-4*a*c))/(2*a)
    if(x1>x2):
        mayor=x1
        return mayor
    else:
        mayor=x2
        return mayor

def funcion_1(x):
    fx=a*math.pow(x,2)+b*x+c
    return fx

while (x<10):
    aux.append(x)
    fxG=a*math.pow(x,2)+b*x+c
    funcion.append(fxG)
    x+=0.8
    
for i in range(0,4):
    xr=(xl+xu)/2
    raices.append(xr)
    fA=funcion_1(xl)
    fB=funcion_1(xr)
    erpv=math.fabs(((cuadratica()-xr)/cuadratica())*100)
    if(i==0):
        aprox.append("-----")
        ea_prx=100
    else:
        j=i-1
        ea_prx=math.fabs(((raices[i]-raices[j])/raices[i])*100)
        aprox.append(ea_prx)
    tabla.append([i+1,xr, str(erpv)+" %",str(aprox[i])+" %"])
    if ((fA*fB)<0):
        xu=xr
    if((fA*fB)>0):
         xl=xr

plt.title("Raiz De La Funcion")
plt.axis([0, 10, -2, 10])
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(0, color="black")
plt.plot(aux, funcion, 'b-', marker='*')
plt.plot(cuadratica(),0, 'cs', label='Raiz Exacta=6.405125')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

print "La raiz mayor es: x= %f\n"%(cuadratica())

print tabulate(tabla, headers=["Iteracion", "Raiz", "Error relativo porcentual verdadero", "Error aproximado"], tablefmt="pipe")
```
